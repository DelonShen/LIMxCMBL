```python
# CCAT-prime
def Gamma_z(_z):
    return (cu.c * 
            (ccl.comoving_radial_distance(cosmo, 1./(1+_z))*u.Mpc)**2
            / (h*100*ccl.background.h_over_h0(cosmo,1./(1+_z)) * (u.km/u.s/u.Mpc)))
 
PeI_intuit = (Gamma_z(3.5)/(40 * u.Gpc**3)
              * 2.22e4 * u_f) 

a,b = CCAT_prime_Pei()
print(b[0].to(u_f))
(PeI_intuit - b)/PeI_intuit
```
