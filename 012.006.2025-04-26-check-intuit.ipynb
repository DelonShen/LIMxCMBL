# SPHEREX

```python
_e = 'SPHEREx'
fids = {
    'R': 41,
    'Omegasurv': experiments[_e]['Omega_field'].value,
    'Omegapix': experiments[_e]['Omega_pix'].value,
    'nPixEff': 2,
}
fids_unit={
    'R': 1,
    'Omegasurv': experiments[_e]['Omega_field'].unit,
    'Omegapix': experiments[_e]['Omega_pix'].unit,
    'nPixEff': 1,
}

scalings = {
    'R': -1,
    'Omegasurv': 0,
    'Omegapix': -1,
    'nPixEff': -1,
}

fid_Pei = 4.1

for _param in fids_unit:
    fid = fids[_param]
    scaling = scalings[_param]
    params = np.linspace(fid/2, fid*2) * fids_unit[_param]
    print(fid, fids_unit[_param])
    args = [{_param:p} for p in params]
    vals = [Pei_dict[_e](**arg) for arg in args ]
    vals = [(val[1][499].to(u_f)/u_f).si for val in vals]
    plt.figure(dpi = 100)
    plt.title(_param)
    plt.plot(params, vals)
    plt.plot(params, fid_Pei*(params/(fid * fids_unit[_param]))**scaling, 'r--')
    plt.show()
```

```python
nuobs, fid_Pei_2 = Pei_dict[_e]()
nurest =nu_Lya
plt.figure(dpi = 100)
plt.title('nuobs and Gamma nuobs')
plt.plot(nuobs, (fid_Pei_2/u_f).si)
plt.plot(nuobs, fid_Pei
         /(nuobs/(335 * u.THz)).si
         *(Gamma_nu(nuobs, nurest)/(29.2 * u.Gpc**3)).si, 'r--')
plt.show()
```
